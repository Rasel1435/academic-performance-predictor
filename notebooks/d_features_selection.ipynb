### Select Best Features For Our Model

we will keep features with an absolute correlation higher than 0.05 (or 5%).

````python
target = 'exam_score'

# Get absolute correlation values for the target
correlations = df.corr()[target].abs()

# Filter features: Keep those with correlation > 0.05
# We always keep the target itself!
selected_features = correlations[correlations > 0.05].index.tolist()

# Create the final dataframe with only selected columns
df_final = df[selected_features]

print(f"Original feature count: {df.shape}")
print(f"Selected feature count: {df_final.shape}")
print(f"\nSelected Columns:\n{df_final.columns}")
````

In the context of our feature selection, the ```.abs()``` function stands for **absolute value**. Its job is to remove the negative sign ($-$) from any number, making every value positive.

**The Mathematical Work**

Mathematically, ```.abs()``` performs the following transformation:

-   $0.83 \rightarrow 0.83$ (Positive stays positive)

-   $-0.17 \rightarrow 0.17$ (Negative becomes positive)

-   $0.00 \rightarrow 0.00$

</br>

In Machine Learning, we are looking for **predictive power**, which is the "strength" of a relationship. We don't care if the relationship is positive or negative; we just care if it is strong (far away from zero).

By using ```.abs()```, we are telling Python:

-   "Show me how much these variables impact the score, regardless of whether they pull the score up or push it down."

</br>

**Summary:** ```.abs()``` turns the "direction" of the relationship into "pure strength" so we can filter all important features with one simple rule.

### This is for ```src/step/d_featuresSelection.py```

````python
def features_selection(df: pd.DataFrame) -> pd.DataFrame:
    target = 'exam_score'

    # Get absolute correlation values for the target
    correlations = df.corr()[target].abs()

    # Filter features: Keep those with correlation > 0.05
    # We always keep the target itself!
    selected_features = correlations[correlations > 0.05].index.tolist()

    # Create the final dataframe with only selected columns
    df_final = df[selected_features]

    print(f"Original feature count: {df.shape}")
    print(f"Selected feature count: {df_final.shape}")
    print(f"\nSelected Columns:\n{df_final.columns}")

    return df_final
````
